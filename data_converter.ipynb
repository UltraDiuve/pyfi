Utilitaire qui permet de lire les trades d'une paire :

```python
def get_trades(pair, data_path=data_path / 'trades'):
    if pair[-4:] == '.csv':
        pair = pair[:-4]
    trades = pd.read_csv(
        data_path / (pair + '.csv'),
        names=['timestamp', 'price', 'volume'],
    ).assign(datetime=lambda x: x['timestamp'].apply(lambda y: dt.datetime.fromtimestamp(y)))
    return(trades)
```

```python
get_trades('XBTAUD').dtypes
```

Utilitaire qui permet de transformer un fichier de trades en un OHLC :

```python
def ohlc_from_trades(trades, freq='10s', datetime_col='datetime'):
    return(
        trades
        .resample(freq, on=datetime_col)
        .agg({
            'price': ['first', 'max', 'min', 'last'],
            'volume': ['sum', 'count'],
        })
        .dropna()
        .reset_index()
        .assign(timestamp=lambda x: x['datetime'].apply(lambda y: int(dt.datetime.timestamp(y))))
        .set_axis(['datetime', 'open', 'high', 'low', 'close', 'volume', 'trade_count', 'timestamp'], axis=1)        
    )
```

```python
ohlc_from_trades(get_trades('XBTAUD'))
```

```python
def kraken_formatted_ohlc_from_trades(trades, freq='10s'):
    return(
        ohlc_from_trades(trades, freq=freq)[
            [
                'timestamp',
                'open', 
                'high',
                'low',
                'close',
                'volume',
                'trade_count',
            ]
        ]
    )
```

```python
kraken_formatted_ohlc_from_trades(get_trades('XBTAUD'))
```

Utilitaire qui permet d'aller chercher l'ohlc d'une paire (si elle existe sur le disque, chargement, sinon calcul, persistage, puis chargement).

Attention ! La frequence doit être en secondes.

```python
pd.read_csv(
                data_path / 'ohlc' / f'XBTAUD_500sec.csv',
                names=['timestamp', 'open', 'high', 'low', 'close', 'volume', 'trade_count']
            )
```

```python
def get_ohlc(pair, int_freq=60, compute_datetime=True):
    
    try:
        ohlc = (
            pd.read_csv(
                data_path / 'ohlc' / f'{pair}_{int_freq}sec.csv',
                names=['timestamp', 'open', 'high', 'low', 'close', 'volume', 'trade_count']
            )
        )
#         print('gotten from file')
    except FileNotFoundError:
#         print('getting from trades')
        ohlc = kraken_formatted_ohlc_from_trades(get_trades(pair), freq=f'{int_freq}s')
#         print('saving to disk')
        ohlc.to_csv(
            data_path / 'ohlc' / f'{pair}_{int_freq}sec.csv',
            index=False,
            header=False,
        )
    if compute_datetime:
        ohlc = ohlc.assign(datetime=lambda x: x['timestamp'].apply(lambda y: dt.datetime.fromtimestamp(y)))
    return(ohlc)
    
```

```python
get_ohlc('XBTAUD', int_freq=500)
```
